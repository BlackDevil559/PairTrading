```python
import yfinance as yf
import pandas as pd
```

```python
closing_price = pd.read_csv('assets\Closing_price_upto_2022.csv')
closing_price.head()
```

```python

import numpy as np
prices = closing_price.set_index('Ticker').transpose()
prices.index = pd.to_datetime(prices.index)
prices.head()
```

```python
daily_returns = prices.pct_change()
log_returns = np.log(prices) - np.log(prices.shift(1))
volatility = daily_returns.std() * np.sqrt(252)
volatility.head()
```

```python
market_ticker = '^BSESN'
market_data = yf.download(market_ticker, start='2012-01-01', end='2022-12-31')
market_prices = market_data['Close']
market_returns = market_prices.pct_change()
market_returns = market_returns.dropna()
print(market_returns.head())
```

```python
import statsmodels.api as sm

common_dates = prices.index.intersection(market_returns.index)
prices = prices.loc[common_dates]
market_returns = market_returns.loc[common_dates]

stock_returns = daily_returns
stock_returns = stock_returns.dropna()
common_dates = stock_returns.index.intersection(market_returns.index)
stock_returns = stock_returns.loc[common_dates]
market_returns = market_returns.loc[common_dates]

beta_dict = {}

for ticker in stock_returns.columns:
    df = pd.concat([stock_returns[ticker], market_returns], axis=1).dropna()
    df.columns = ['stock', 'market']

    X = sm.add_constant(df['market'])
    y = df['stock']

    model = sm.OLS(y, X).fit()

    beta_dict[ticker] = model.params['market']

beta_series = pd.Series(beta_dict)
print(beta_series)
```

```python
prices.dropna()
market_prices.dropna()
stock_returns = np.log(prices) - np.log(prices.shift(1))
market_returns = np.log(market_prices) - np.log(market_prices.shift(1))
window = 250

def rolling_beta(stock_ret, market_ret, window):
    def beta_func(x):
        m_returns = market_ret.loc[x.index].to_numpy().flatten()
        s_returns = x.to_numpy().flatten()
        beta = np.polyfit(m_returns, s_returns, 1)[0]
        return beta

    return stock_ret.rolling(window).apply(beta_func, raw=False)
```

```python
common_dates = stock_returns.index.intersection(market_returns.index)
stock_returns = stock_returns.loc[common_dates]
market_returns = market_returns.loc[common_dates]
print(stock_returns.shape)
print(market_returns.shape)
```

```python
rolling_betas = {}
for ticker in stock_returns.columns:
    rolling_betas[ticker] = rolling_beta(stock_returns[ticker], market_returns, window)

rolling_beta_df = pd.DataFrame(rolling_betas)
print(rolling_beta_df.tail())
rolling_volatility = stock_returns.rolling(30).std()
ma_50 = prices.rolling(50).mean()
ma_200 = prices.rolling(200).mean()
```

```python
import yfinance as yf
import pandas as pd
import time
import random

tickers = prices.columns.tolist()
fundamental_data_list = []

keys = {
    'marketCap': 'MarketCap',
    'trailingPE': 'TrailingPE',
    'forwardPE': 'ForwardPE',
    'trailingEps': 'TrailingEPS',
    'forwardEps': 'ForwardEPS',
    'totalRevenue': 'TotalRevenue',
    'dividendYield': 'DividendYield',
    'beta': 'Beta', 
    'sector': 'Sector',
    'industry': 'Industry',
    'fullTimeEmployees': 'FullTimeEmployees'
}

def fetch_fundamentals(ticker):
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        data = {'Ticker': ticker}
        for key, col_name in keys.items():
            data[col_name] = info.get(key, None)
        return data
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return None

def chunker(seq, size):
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))

batch_size = 10 
for batch in chunker(tickers, batch_size):
    for ticker in batch:
        result = fetch_fundamentals(ticker)
        if result:
            fundamental_data_list.append(result)
    sleep_time = random.uniform(2, 5) 
    print(f"Processed batch of {len(batch)} tickers. Sleeping for {sleep_time:.2f} seconds...")
    time.sleep(sleep_time)

df_fundamentals = pd.DataFrame(fundamental_data_list)
print(df_fundamentals.head())
```

```python
last_date = prices.index.max()
data = {}

for ticker in prices.columns:
    # 1. Last price

    # 2. Annualized volatility (from daily returns)
    vol = volatility.get(ticker, np.nan)

    # 3. Static beta from the regression (if available)
    beta_static = beta_series.get(ticker, np.nan)

    # 4. Latest rolling beta (use the last non-NaN value)
    rb_series = rolling_beta_df[ticker].dropna()
    last_rolling_beta = rb_series.iloc[-1] if not rb_series.empty else np.nan

    # 5. Latest rolling volatility (30-day volatility)
    rvol_series = rolling_volatility[ticker].dropna()
    last_rolling_vol = rvol_series.iloc[-1] if not rvol_series.empty else np.nan

    # 6. Latest 50-day moving average
    ma50_series = ma_50[ticker].dropna()
    last_ma50 = ma50_series.iloc[-1] if not ma50_series.empty else np.nan

    # 7. Latest 200-day moving average
    ma200_series = ma_200[ticker].dropna()
    last_ma200 = ma200_series.iloc[-1] if not ma200_series.empty else np.nan

    # Store all features for this ticker in a dictionary.
    data[ticker] = {
        # 'last_price': last_price,
        'volatility': vol,
        'beta_static': beta_static,
        'rolling_beta': last_rolling_beta,
        'rolling_volatility': last_rolling_vol,
        'ma_50': last_ma50,
        'ma_200': last_ma200
    }

features_df = pd.DataFrame.from_dict(data, orient='index')
print(features_df.head())
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
df_fundamentals_indexed = df_fundamentals.set_index('Ticker')
if 'Sector' in df_fundamentals_indexed.columns:
    df_fundamentals_indexed['Sector'] = df_fundamentals_indexed['Sector'].fillna('Unknown')
    df_fundamentals_indexed['Industry'] = df_fundamentals_indexed['Industry'].fillna('Unknown')
    le = LabelEncoder()
    df_fundamentals_indexed['Sector_Encoded'] = le.fit_transform(df_fundamentals_indexed['Sector'])
    df_fundamentals_indexed['Industry_Encoded'] = le.fit_transform(df_fundamentals_indexed['Industry'])
```

```python
price_features_path = 'assets/price_features.csv'
fundamental_features_path = 'assets/fundamental_features.csv'

df_fundamentals_indexed.to_csv(fundamental_features_path, index=True)
features_df.to_csv(price_features_path, index=True)

features_df.to_csv(fundamental_features_path , index=True)
df_fundamentals_indexed.to_csv(price_features_path, index=True)
```
